```python
def cov(x,y):
    return np.mean(x*y)-np.mean(x)*np.mean(y)
```

```python
def cor(x,y):
    return cov(x,y)/sqrt(np.var(x)*np.var(y))
```

```python
print(cov(nba_data['APG'],nba_data['PPG']))
print(np.cov(nba_data['APG'],nba_data['PPG'])[0,1])
print('------------------------------------')
print(cor(nba_data['APG'],nba_data['PPG']))
print(np.corrcoef(nba_data['APG'],nba_data['PPG'])[0,1])
```

```python
def mykmeans(X,k,c=None,distance=(lambda x,y: np.linalg.norm(x-y))):
    print('Kmeans learning start')
    X=np.array(X)
    if not c:
        c=X[np.random.choice(len(X),size=k,replace=False)]
    res=[0] * len(X)
    for iter in range(100):
        groups=[ []for _ in range(k)]
        distance_sum=0
        if iter%20==19:
            print(f'{iter+1}th iteration')
        for i,player in enumerate(X):
            distances=[distance(player,center) for center in c]
            group_num=np.argmin(distances)
            distance_sum+=distances[group_num]
            res[i]=group_num
            groups[group_num].append(player)
        stop=True
        for i, group in enumerate(groups):
            new_center=np.array(group).mean(axis=0)
            if np.linalg.norm(c[i]-new_center)>0.001:
                stop=False
            c[i] = new_center
        if stop:
            print(f'Early stop at {iter+1}th iteration')
            break
    return res, c, distance_sum/len(X), iter+1
```

```python
mykmeans(nba_data,20)
```

```python
def check_kmeans(clust_res, pos_list, logging=False):
    assert(len(clust_res)==len(pos_list))
    pos_value_counts=pos_list.value_counts()
    k=len(pos_value_counts)
    true_positive=0
    already_assign=[]
    for i in range(k):
        pos_i=pos_value_counts.index[np.argmax(pos_value_counts)]
        if logging: print(pos_i)
        num_groups_with_pos_i=[0]*k
        for j,position in enumerate(pos_list):
            if position==pos_i and not clust_res[j] in already_assign :
                num_groups_with_pos_i[clust_res[j]]+=1
        if logging: print(num_groups_with_pos_i)
        assign_cluster=np.argmax(num_groups_with_pos_i)
        true_positive+=num_groups_with_pos_i[assign_cluster]
        already_assign.append(assign_cluster)
        pos_value_counts=pos_value_counts.drop(labels=[pos_i])
    return true_positive/len(clust_res)
```

```python
def test_kmeans(data,k,pos,distance=None,num=10):
    max_acc=0
    sum_early_stop=0
    for __ in range(num):
        if distance:
            res, _, _, early_stop = mykmeans(data,k,distance=distance)
        else:
            res, _, _ , early_stop= mykmeans(data,k)
        sum_early_stop+=early_stop
        acc=check_kmeans(res, pos)
        if max_acc < acc:
            max_acc=acc
    return max_acc, sum_early_stop/num
```
